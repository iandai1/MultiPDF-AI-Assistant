```python
from langchain_openai import OpenAIEmbeddings
import pandas as pd
```

```python
from langchain.evaluation import load_evaluator
import numpy as np
from scipy.spatial.distance import cosine
```

```python
OPENAI_API_KEY = ""
```

```python
def get_embeddings_for_words():
    embeddings = OpenAIEmbeddings(
        model="text-embedding-ada-002",
        openai_api_key=OPENAI_API_KEY
    )
    return embeddings

embedding_function = get_embeddings_for_words()
test_vector = embedding_function.embed_query("dog")
print(test_vector)
print("długość wektora:",len(test_vector))
```

```python
from langchain.evaluation import load_evaluator

evaluator = load_evaluator(evaluator="embedding_distance",
                            embeddings=embedding_function)

evaluator.evaluate_strings(prediction="Warsaw", reference="Poland")
```
